```python
import matplotlib.pyplot as plt
import math
import random
import numpy as np
```

```python
class Coronavirus:
    def __init__(self, n, p, e, x_p, pop):
        self.n = n
        self.p = p
        self.e = e
        self.x_p = x_p
        self.pop = pop
    
    def growthcalc(self):
        ls = []
        ls.append(self.n)
        while(self.n <= self.pop):
            self.n = round(self.n*(1-self.x_p)*(1 + self.e * self.p))
            ls.append(self.n)
        return ls
    
    def simulate(self):
        largels = []
        largetup = []
        ##probability changing simulation
        lsi = list(np.linspace(0.1, 1, 100))
        for i in range(len(lsi)):
            for j in range(len(lsi)):
                for k in range(1, 100):
                    ls = []
                    ls.append(self.n)
                    while(self.n <= self.pop):
                        self.n = round(self.n*(1-lsi[j])*(1 + k * lsi[i]))
                        ls.append(self.n)
                    largels.append(ls)
                    largetup.append((self.n, lsi[j], lsi[i], k))
        return largels, largetup
```

```python
def plotter(x, y):
    plt.xlabel('Days')
#     plt.yscale('log')
    plt.ylabel('Number Infected')
    plt.title('Regular Graph')
    plt.plot(x, y)
    plt.show()
```

```python
def logplotter(x, y):
    plt.xlabel('Days')
#     plt.yscale('log')
    plt.ylabel('Number Infected')
    plt.title('Logarithmic Scale')
    plt.yscale('log')
    plt.plot(x, y)
    plt.show()
```
